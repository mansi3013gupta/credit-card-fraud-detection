DATA Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_csv('creditcard.csv')  # Replace with your dataset path

# Set the style for seaborn
sns.set_style('whitegrid')

# Visualization 1: Class Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=df, palette='coolwarm')
plt.title('Class Distribution')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Count')
plt.show()

```

```python

# Visualization 2: Correlation Heatmap
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap')
plt.show()
```

```python

# Visualization 3: Transaction Amount Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Amount'], kde=True, bins=50, color='blue')
plt.title('Transaction Amount Distribution')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.show()

```

```python
# Visualization 4: Fraud vs Non-Fraud Amount
plt.figure(figsize=(8, 5))
sns.boxplot(x='Class', y='Amount', data=df, palette='Set2')
plt.title('Transaction Amount by Class')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Transaction Amount')
plt.show()

```

```python
# Visualization 5: PCA Features Distribution
# Assuming the dataset has PCA components (e.g., V1, V2, ..., V28)
pca_features = [col for col in df.columns if 'V' in col]

plt.figure(figsize=(12, 8))
for feature in pca_features[:5]:  # Visualizing first 5 PCA features
    sns.kdeplot(df[feature], label=feature, fill=True)
plt.title('Distribution of PCA Features')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()

```

```python
# Visualization 6: Time vs Transaction Amount
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Time', y='Amount', hue='Class', data=df, palette='Set1', alpha=0.5)
plt.title('Time vs Transaction Amount')
plt.xlabel('Time (Seconds from Start)')
plt.ylabel('Transaction Amount')
plt.legend(title='Class', loc='upper right')
plt.show()

```

```python
# Visualization 7: Fraud Proportion in Time Intervals
df['Hour'] = (df['Time'] // 3600) % 24  # Convert 'Time' to hours
hourly_fraud = df.groupby('Hour')['Class'].mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=hourly_fraud.index, y=hourly_fraud.values, palette='coolwarm')
plt.title('Fraud Proportion by Hour of the Day')
plt.xlabel('Hour of Day')
plt.ylabel('Proportion of Fraudulent Transactions')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
df=pd.read_csv('/content/creditcard.csv')


# Separate features (X) and target variable (y)
X = df.drop('Class', axis=1)
y = df['Class']

# 1. Handle Class Imbalance

# Option 1: Oversampling with SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# 2. Feature Scaling (Normalize the features)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)

# 3. Data Splitting (80-20 split)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)

# Output the results
print(f"Original Dataset Shape: {X.shape}, {y.shape}")
print(f"Resampled Dataset Shape: {X_resampled.shape}, {y_resampled.shape}")
print(f"Training Set Shape: {X_train.shape}, {y_train.shape}")
print(f"Test Set Shape: {X_test.shape}, {y_test.shape}")
```

LOGISTIC REGRESSION MODEL

```python
# Import necessary libraries for modeling and evaluation
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# 1. Initialize Logistic Regression Model
logreg = LogisticRegression(random_state=42)

# 2. Train the model on the training data
logreg.fit(X_train, y_train)

# 3. Make predictions on the test set
y_pred = logreg.predict(X_test)

# 4. Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# 5. Display the results
print("Model Evaluation Metrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")

# 6. Confusion Matrix (optional)
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Non-Fraud', 'Fraud'], yticklabels=['Non-Fraud', 'Fraud'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
